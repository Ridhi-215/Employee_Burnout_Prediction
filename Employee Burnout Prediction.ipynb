```python
data = data.drop(columns=['EmpID'])
data = data.drop(columns=['FirstName'])
data = data.drop(columns=['LastName'])
data = data.drop(columns=['Supervisor'])
data = data.drop(columns=['ADEmail'])
data = data.drop(columns=['DOB'])
data = data.drop(columns=['LocationCode'])
data = data.drop(columns=['TerminationType'])
data = data.drop(columns=['TerminationDescription'])
data = data.drop(columns=['State'])
```

```python
# Calculating 'Days Worked' from 'StartDate'
data['StartDate'] = pd.to_datetime(data['StartDate'])
data['Days Worked'] = (pd.Timestamp.now() - data['StartDate']).dt.days
```

```python
data=data.drop(columns=['StartDate'])
```

```python
data.shape
```

```python
#Encoding categorical values to numerical vales
from sklearn.preprocessing import LabelEncoder           #import library
encoder=LabelEncoder()            #create object
data['Title']=encoder.fit_transform(data['Title'])   
data['EmployeeStatus']=encoder.fit_transform(data['EmployeeStatus'])   
data['EmployeeType']=encoder.fit_transform(data['EmployeeType'])        
data['PayZone']=encoder.fit_transform(data['PayZone'])           
data['EmployeeClassificationType']=encoder.fit_transform(data['EmployeeClassificationType'])
data['DepartmentType']=encoder.fit_transform(data['DepartmentType'])                   
data['Division']=encoder.fit_transform(data['Division'])
data['JobFunctionDescription']=encoder.fit_transform(data['JobFunctionDescription'])
data['GenderCode']=encoder.fit_transform(data['GenderCode'])                  
data['RaceDesc']=encoder.fit_transform(data['RaceDesc'])
data['MaritalDesc']=encoder.fit_transform(data['MaritalDesc'])   
data['Performance Score']=encoder.fit_transform(data['Performance Score'])
```

```python
data.head(10)
```

```python
# Normalizing 'Burnout' column
from sklearn.preprocessing import MinMaxScaler

x = data['Burnout'].values.reshape(-1, 1)  # Reshape to 2D array

scaler = MinMaxScaler()  
x_scaled = scaler.fit_transform(x) 

data['Burnout'] = x_scaled
```

```python
data
```

```python
# Finding Correlation with the 'Burnout' Column
important_columns = ['Days Worked', 'Performance Score', 'Current Employee Rating', 'Title', 'EmployeeStatus', 'PayZone','EmployeeType','EmployeeClassificationType','DepartmentType','Division','JobFunctionDescription','GenderCode','MaritalDesc','RaceDesc']
correlation_matrix = data[important_columns + ['Burnout']].corr()

burnout_correlation = correlation_matrix['Burnout'].sort_values(ascending=False)
print(burnout_correlation)
```

```python
# List of columns to drop based on correlation analysis
columns_to_drop = [
    'Title', 
    'MaritalDesc', 
    'JobFunctionDescription', 
    'Current Employee Rating', 
    'Performance Score', 
    'GenderCode', 
    'Division', 
    'DepartmentType', 
    'EmployeeClassificationType'
]
data = data.drop(columns=columns_to_drop)
```

```python
import seaborn as sns
features = ['EmployeeStatus', 'EmployeeType', 'PayZone', 'RaceDesc', 'Days Worked']
plt.figure(figsize=(14, 10))

for i, feature in enumerate(features):
    plt.subplot(2, 3, i + 1)
    sns.regplot(x=feature, y='Burnout', data=data, scatter_kws={'s':50}, line_kws={'color':'red'})
    plt.title(f'Burnout vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('Burnout')
plt.tight_layout()
plt.show()
```

```python
y = data['Burnout']
X = data.drop('Burnout', axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train), index=X_train.index, columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), index=X_test.index, columns=X_test.columns)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

linear_regression_model = LinearRegression()

linear_regression_model.fit(X_train, y_train)

y_pred = linear_regression_model.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate root mean squared error
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("Root Mean Squared Error:", rmse)

# Calculate mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```
